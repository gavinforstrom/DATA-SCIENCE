```python
list_of_names = ['Gavin' , 'Nate' , 'Brigham']

compare_names = (names.query('name is in @list_of_names')
    .groupby('name')
    .mean()
    .filter(['AK', 'Total'])
    .sort_values('AK' , ascending = False))

compare_names    
```
